# quickstart

Quickstart example using pyglotaran simulated sequential spectral decay data

```
from __future__ import annotations

from glotaran.io import save_result
from glotaran.optimization.optimize import optimize
from glotaran.project.scheme import Scheme
from glotaran.testing.simulated_data import sequential_spectral_decay
from pyglotaran_extras import plot_data_overview
from pyglotaran_extras.plotting.plot_overview import plot_overview
```

## Load simulated data

The dataset is generated by pyglotaran's built-in sequential spectral decay simulator —
no external files needed.

```
dataset = sequential_spectral_decay.DATASET
plot_data_overview(dataset, nr_of_data_svd_vectors=3, linlog=True, linthresh=1);
```

## Define model and parameters

The model (`models/model.yaml`) specifies a 3-compartment sequential decay with a
Gaussian IRF. The parameters (`models/parameters.yaml`) supply initial guesses for
the three kinetic rates and the IRF center/width.

```
scheme = Scheme(
    model="models/model.yaml",
    parameters="models/parameters.yaml",
    maximum_number_function_evaluations=99,
    data={"my_data": dataset},
)
scheme.validate()
```

## Optimize

```
result = optimize(scheme, raise_exception=True)
result
```

## Results

Optimized parameters, overview plot, and saved result.

```
result.optimized_parameters
```

```
plot_overview(result.data["my_data"], linlog=True, linthresh=1);
```

```
save_result(result=result, result_path="results", format_name="folder", allow_overwrite=True)
```
